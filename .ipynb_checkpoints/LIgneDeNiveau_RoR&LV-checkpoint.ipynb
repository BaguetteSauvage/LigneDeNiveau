## Positionnement du problème

&emsp; On se donne une fonction $f$ définie sur un domaine $\mathcal{D}$ de $\mathbb{R}^{2}$, et l'on cherche une ligne de niveau où $f$ prend toujours la valeur $c$, c'est-à-dire un ensemble $\lbrace (x,y) \in \mathbb{R}^{2}, f(x,y) = c \rbrace$.

&emsp; On va d'abord ramener le problème à la recherche de telles courbes de niveau sur le carré unité, puis on appliquera cette méthode à un domaine carré quelconque grâce à une composition par une fonction affine. 

## La fonction find_seed

&emsp; Pour initialiser la recherche d'une ligne de niveau, il faut commencer par en trouver un point. Comme on a ramené le problème au carré unité, on va rechercher un point $x_{0}$ sur l'arrête gauche du carré unité tel que $\vert f(x_{0}) - c \vert \leq \varepsilon$, où $\varepsilon$ est laissé au choix de l'utilisateur, avec une valeur par défaut.

&emsp; Pour assurer que ce $x_{0}$ existe, on peut utiliser le théorème des valeurs intermédiaires sur le segment $[0,1]$. En effet, si $c$ est compris entre $f(0)$ et $f(1)$, ce qui est équivalent à $(f(0) - c)(f(1) -c) \leq 0$, alors on assure l'existence de $x_{0}$. 
On recherche ensuite la valeur $x_{0}$ au moyen d'une dichotomie, qui est assurée de converger selon nos hypothèses précédentes. 

&emsp; On a donc écris un programme qui appliquait ce principe, qui prend en argument $f$, $c$, éventuellement $a$, $b$ au cas où la dichotomie n'est pas sur $[0,1]$ et $\varepsilon$ si l'on souhaite changer la précision.  

```python
def find_seed(f, c, a = 0, b = 1, eps = 10**(-10)):
    fa, fb = f(a), f(b)
    if (fa - c)*(fb - c) > 0:
        return False
    ec = min(abs(fa-c), abs(fb-c))
    while ec > eps:
        m = (a+b)/2
        fm = f(m)
        if (fa - c)*(fm - c) <= 0:
            b = m
            fb = fm
        else:
            a = m
            fa = fm
        ec = min(abs(fa-c), abs(fb-c))
    if ec == abs(fa-c):
        return a
    return b
```

## Simple contour

&emsp; On suppose maintenant avoir $x_{0}$ comme décris précédemment. On va maintenant chercher un fragment de la courbe de niveau de $f$ sur le carré unité, en discrétisant cette courbe sur un ensemble de points, deux points adjacents séparés environ de $\delta$.

&emsp; On va commencer par essayer d'approximer la courbe de niveau par sa tangeante. Si l'on a les hypothèses du théorème des fonctions implicites, on peut localement (au voisinnage de $x_{0}$) décrire la courbe de niveau comme une courbe paramétrée $\mathcal{C}^{1}$, on a donc l'existence d'une tangeante en ce point. Cette tangeante coupe le cercle de centre $x_{0}$ et de rayon $\delta$ en deux points. On choisit le point le plus éloigné du point qui précédait $x_{0}$ sur la  courbe, s'il existe. Sinon, $x_{0}$ est le premier point trouvé, et seul l'un des deux points est dans le carré unité. Cette convention permet de ne pas voyager entre deux de la ligne de niveau lors du tracé.

&emsp; On passe ensuite à l'étape finale de la recherche d'un nouveau point, qui se fait par la méthode de Newton. La méthode de la tangeante permet de se placer sur un point $t$ relativement proche d'un point $x_{1}$ qui vérifie $\vert f(x_{1}) - c \vert \leq \varepsilon$. On part de ce point $t$ et l'on applique la méthode de Newton modifiée selon l'axe des $x$ ou des $y$ por trouver le point $x_{1}$. Comme $t$ est relativement proche de $x_{1}$, cette méthode a de fortes chances de converger. Quand au choix de l'axe des $x$ ou des $y$, on le fait selon la pente de la tangeante: si elle est supérieure à 1, la tangeante est plutôt vertiacle, et la plus petite erreur est selon l'axe des $x$. Sinon, la tangeante est plutôt verticale, et l'erreur est selon l'axe des $y$.

```python
def simple_contour(f, c , delta = 0.01):
    tx,ty = [],[]
    y0 = find_seed(f, c)
    tx.append(0)
    ty.append(y0)
    continuer, compteur = True, 1000 #on dessine au maximum 1000 points
    while continuer and compteur > 0:
        (x, y) = tx[-1], ty[-1]
        grad_f = gradient(f, x, y)
        tangeante = -(grad_f[0]/grad_f[1])
        alpha = atan(tangeante)
        (nx, ny) = (x+delta*cos(alpha), y+delta*sin(alpha))
        (mx,my) = (x-delta*cos(alpha), y-delta*sin(alpha))
        if len(tx)>1:
            x,y = tx[-2],ty[-2]
            if (x-nx)**2+(y-ny)**2 > (x-mx)**2+(y-my)**2:
                A = nx
                B = ny
            else:
                A = mx
                B = my
            if abs(tangeante) < 1:
                A2 , B2 = newtonx(f,A,B,c,epsilon = 2**(-48))
            else:
                A2 , B2 = newtony(f,A,B,c,epsilon = 2**(-48))
        else:
            A2 = nx
            B2 = ny
        tx.append(A2)
        ty.append(B2)
        compteur -= 1
        if tx[-1] > 1 or tx[-1]< 0 or ty[-1] > 1 or ty[-1]< 0:
            continuer = False #si la fonction sort du carré, on arrête
    return tx , ty

def newtonx(g,A,B,c,epsilon):
    f = lambda x,y: g(x,y) - c
    if abs(f(A,B)) < epsilon:
        return A,B
    else:
        compteur = 100 #c'est un nombre suffisant d'essais car on est proche du point que l'on cherche
        while abs(f(A,B)) > epsilon and compteur:
            C = gradient(f,A,B)[1]
            B = B - f(A,B)/C
            compteur -=1
    return A , B

def newtony(g,A,B,c,epsilon):
    f = lambda x,y: g(x,y) - c
    if abs(f(A,B)) < epsilon:
        return A,B
    else:
        compteur = 100
        while abs(f(A,B)) > epsilon and compteur:
            C = gradient(g,A,B)[0]
            A = A - f(A,B)/C
            compteur -=1
    return A , B
```

## Contour complexe

&emsp; Maintenant que l'on a la fonction simple contour, on va écrire la fonction contour complexe qui prend comme argument la fonction $f$, la valeur $c$ de la ligne de niveau et deux tableaux qui permettent de former un quadrillage carré de l'ensemble de définition. On va commencer par faire une fonction find_seed_complet qui suivra le même principe que find_seed et permettra de trouver si possible le départ des lignes de niveaux sur les 4 côtés des carrés du quadrillage.

```python
def find_seed_gauche(f, cste = 0, eps = 2**(-26)): #permet de trouver une 'seed' sur l'arrete gauche de chaque carré
    a,b = 0,1
    f_gauche = lambda x: f(0, x)
    return dichotomy(f_gauche, cste, 0, 1, eps)

def find_seed_bas(f, cste = 0, eps = 2**(-26)): #permet de trouver une 'seed' sur l'arrete du bas de chaque carré
    a,b = 0,1
    f_bas = lambda x: f(x,0)
    return dichotomy(f_bas, cste, 0, 1, eps)

def find_seed_haut(f, cste = 0, eps = 2**(-26)): #permet de trouver une 'seed' sur l'arrete du haut de chaque carré
    a,b = 0,1
    f_haut = lambda x: f(x,1)
    return dichotomy(f_haut, cste, 0, 1, eps)

def find_seed_droite(f, cste = 0, eps = 2**(-26)): #permet de trouver une 'seed' sur l'arrete droite de chaque carré
    a,b = 0,1
    f_haut = lambda x: f(1,x)
    return dichotomy(f_haut, cste, 0, 1, eps)

def find_seed_complet(f,c,k): #la variable k permet de savoir de quel côté est traîté par la fonction
    if k == 0:
        return 0,find_seed_gauche(f, c)
    if k == 1:
        return 1,find_seed_droite(f,c)
    if k == 2:
        return find_seed_bas(f,c),0
    if k ==3:
        return find_seed_haut(f,c),1
```

&emsp; Avec cette fonction find_seed_complet, il ne reste plus qu'à modifier la fonction $f$ pour que les différents ensembles définis par les carrés du quadrillage correspondent à l'ensemble $[0,1]$x$[0,1]$, puis à appliquer la fonction contour simple et remodifier les valeurs pour qu'elles correspondent au carré dans lequel elles ont été trouvées.

```python
def contour(f, xc=[0.0, 1.0], yc=[0.0, 1.0], c=0.0, delta=0.01):
    n_x, n_y = len(xc)-1, len(yc)-1
    cote = xc[1]-xc[0] #cote est la longueur des côtés des différents carrés du quadrillage
    xs, ys = [], []
    color1,color2,color3 = rd.random(),rd.random(),rd.random() #color1-2-3 permettent de définir la couleur de la ligne de niveaux
    
    for i in range(n_x):
        for j in range(n_y):
            def g(x,y):  #on définit la fonction modifiée pour appliquer contour simple
                return f(xc[i] + cote*x,yc[j] + cote*y)
            for k in range(4):
                Lint_x = [xc[i] + (xc[i+1]-xc[i])*x for x in simple_contour(g, c, delta,k)[0]]
                Lint_y = [yc[j] + (yc[j+1]-yc[j])*y for y in simple_contour(g, c, delta,k)[1]]
                xs.append(Lint_x)
                ys.append(Lint_y)
    for x,y in zip(xs,ys):
        plt.plot(x,y,"-",color = (color1,color2,color3))
```

&emsp; On peut par ailleurs faire un fonction quadrillage qui permet de renvoyer les 2 tableaux xc et yc qui forment le quadrillage.

```python
def quadrillage(x1,x2,y1,y2,c):
    '''fait un quadrillage de l'ensemble [x1;x2]x[y1;y2]'''
    xc = [x1]
    yc = [x2]
    while x1+c <= x2:
        xc.append(x1+c)
        x1 += c
    while y1+c <= y2:
        yc.append(y1+c)
        y1 += c
    return xc , yc
```

&emsp; Une fois tout le code réunit on obtient ceci:

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
import random as rd
from scipy.misc import derivative

f = lambda x,y: 2*(np.exp(-x**2 - y**2)-np.exp(-(x-1)**2 - (y-1)**2))

def dichotomy(g, cste, a=0, b=1, eps= 10**(-26)):
    ga, gb = g(a), g(b)
    if (ga - cste)*(gb - cste) > 0:
        return False
    ec = min(abs(ga-cste), abs(gb-cste))
    while ec > eps:
        c = (a+b)/2
        gc = g(c)
        if (ga - cste)*(gc - cste) <= 0:
            b = c
            gb = gc
        else:
            a = c
            ga = gc
        ec = min(abs(ga-cste), abs(gb-cste))
    if ec == abs(ga-cste):
        return a
    return b

def partial_derivative(func, var=0, point=[]):
    args = point[:]
    def wraps(x):
        args[var] = x
        return func(*args)
    return derivative(wraps, point[var], dx = 1e-8)

def gradient(g, x, y):
    return partial_derivative(g,0,[x,y]) , partial_derivative(g,1,[x,y])

def find_seed_gauche(g, cste = 0, eps = 2**(-26)):
    g_gauche = lambda x: g(0, x)
    return dichotomy(g_gauche, cste, 0, 1, eps)

def find_seed_bas(g, cste = 0, eps = 2**(-26)):
    g_bas = lambda x: g(x,0)
    return dichotomy(g_bas, cste, 0, 1, eps)

def find_seed_haut(g, cste = 0, eps = 2**(-26)):
    g_haut = lambda x: g(x,1)
    return dichotomy(g_haut, cste, 0, 1, eps)

def find_seed_droite(g, cste = 0, eps = 2**(-26)):
    g_haut = lambda x: g(1,x)
    return dichotomy(g_haut, cste, 0, 1, eps)

def find_seed_complet(g,c,k):
    if k == 0:
        return 0,find_seed_gauche(g, c)
    if k == 1:
        return 1,find_seed_droite(g,c)
    if k == 2:
        return find_seed_bas(g,c),0
    if k ==3:
        return find_seed_haut(g,c),1

def simple_contour(f, c , delta,k):
    tx,ty = [],[]
    x0,y0 = find_seed_complet(f, c,k)
    if (k==0 or k==1) and not y0:
        return tx,ty
    if (k==2 or k ==3) and not x0:
        return tx,ty
    tx.append(x0)
    ty.append(y0)
    continuer, compteur = True, 1000
    while continuer and compteur > 0:
        (x, y) = tx[-1], ty[-1]
        grad_f = gradient(f, x, y)
        tangeante = -(grad_f[0]/grad_f[1])
        alpha = atan(tangeante)
        (nx, ny) = (x+delta*cos(alpha), y+delta*sin(alpha))
        (mx,my) = (x-delta*cos(alpha), y-delta*sin(alpha))
        if len(tx)>1:
            x,y = tx[-2],ty[-2]
            if (x-nx)**2+(y-ny)**2 > (x-mx)**2+(y-my)**2:
                A = nx
                B = ny
            else:
                A = mx
                B = my
            if abs(tangeante) < 1:
                A2 , B2 = newtonx(f,A,B,c,epsilon = 2**(-48))
            else:
                A2 , B2 = newtony(f,A,B,c,epsilon = 2**(-48))
        else:
            if nx > 0 and nx < 1 and ny < 1 and ny > 0:
                A2 = nx
                B2 = ny
            else:
                A2 = mx
                B2 = my
        tx.append(A2)
        ty.append(B2)
        compteur -= 1
        if tx[-1] > 1 or tx[-1]< 0 or ty[-1] > 1 or ty[-1]< 0:
            continuer = False
    return tx , ty

def newtonx(g,A,B,c,epsilon):
    f = lambda x,y: g(x,y) - c
    if abs(f(A,B)) < epsilon:
        return A,B
    else:
        compteur = 100
        while abs(f(A,B)) > epsilon and compteur:
            C = gradient(f,A,B)[1]
            B = B - f(A,B)/C
            compteur -=1
    return A , B

def newtony(g,A,B,c,epsilon):
    f = lambda x,y: g(x,y) - c
    if abs(f(A,B)) < epsilon:
        return A,B
    else:
        compteur = 100
        while abs(f(A,B)) > epsilon and compteur:
            C = gradient(g,A,B)[0]
            A = A - f(A,B)/C
            compteur -=1
    return A , B

def quadrillage(x1,x2,y1,y2,c):
    xc = []
    yc = []
    while x1+c <= x2:
        xc.append(x1)
        x1 += c
    while y1 <= y2:
        yc.append(y1)
        y1 += c
    return xc , yc

def contour(f, xc=[0.0, 1.0], yc=[0.0, 1.0], c=0.0, delta=0.01):
    n_x, n_y = len(xc)-1, len(yc)-1
    cote = xc[1]-xc[0]
    xs, ys = [], []
    color1,color2,color3 = rd.random(),rd.random(),rd.random()
    for i in range(n_x):
        for j in range(n_y):
            def g(x,y):
                return f(xc[i] + cote*x,yc[j] + cote*y)
            for k in range(4):
                Lint_x = [xc[i] + (xc[i+1]-xc[i])*x for x in simple_contour(g, c, delta,k)[0]]
                Lint_y = [yc[j] + (yc[j+1]-yc[j])*y for y in simple_contour(g, c, delta,k)[1]]
                xs.append(Lint_x)
                ys.append(Lint_y)
    for x,y in zip(xs,ys):
        plt.plot(x,y,"-",color = (color1,color2,color3))


A = np.linspace(-1,1,9) #A est l'ensemble des lignes de niveau que l'on veut tracer
a,b = quadrillage(-2.0,3.0,-2.0,3.0,0.5)
for k in A:
    contour(f,a,b,k)
plt.show()
```
